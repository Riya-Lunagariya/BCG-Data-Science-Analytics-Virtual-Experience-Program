```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# 1. Distribution of Annual Consumption (`cons_12m`)
plt.figure(figsize=(8, 6))
# Apply log transformation to 'cons_12m' to normalize distribution for visualization
# Add 1 to avoid log(0) which is undefined
log_cons_12m = np.log(client['cons_12m'] + 1)
sns.histplot(log_cons_12m, bins=30, kde=True)
plt.title('Log Distribution of Annual Consumption (cons_12m)')
plt.xlabel('Log of Consumption')
plt.ylabel('Frequency')
plt.show()
```

Analysis:

Most customers have a log consumption value between roughly 7 to 11.
There is a peak around the 9 to 10 range on the log scale, which suggests that there is a common range of consumption among a significant number of customers.
The distribution tails off as the log consumption value increases, indicating fewer customers with very high consumption levels.
The long tail to the right (higher log consumption values) suggests that while most customers have lower consumption levels, there are a few with much higher usage.

**Consumption**

Let's see the distribution of the consumption in the last year and month. To perform EDA on the distribution of consumption in the last year (cons_12m) and the last month (cons_last_month), we can use histograms or boxplots to visualize the distributions. Given that these distributions can be highly skewed, as suggested by the descriptive statistics you provided earlier, it's often helpful to apply a log transformation to the data before plotting.

```python
# Apply log transformation to the consumption data to reduce skewness
# Adding a small constant to avoid log(0)
client['log_cons_12m'] = np.log(client['cons_12m'] + 1)
client['log_cons_last_month'] = np.log(client['cons_last_month'] + 1)

# Plotting the distribution of log-transformed consumption in the last year
plt.figure(figsize=(10, 5))
sns.histplot(client['log_cons_12m'], bins=50, kde=True)
plt.title('Log-Transformed Distribution of Consumption in the Last Year (cons_12m)')
plt.xlabel('Log of Consumption in the Last Year')
plt.ylabel('Frequency')
plt.show()

# Plotting the distribution of log-transformed consumption in the last month
plt.figure(figsize=(10, 5))
sns.histplot(client['log_cons_last_month'], bins=50, kde=True)
plt.title('Log-Transformed Distribution of Consumption in the Last Month (cons_last_month)')
plt.xlabel('Log of Consumption in the Last Month')
plt.ylabel('Frequency')
plt.show()
```

Box plot for the above analysis

```python
# Apply a log transformation to the consumption columns to reduce skewness
# Adding a small constant to avoid log(0)
client['log_cons_12m'] = np.log(client['cons_12m'] + 1)
client['log_cons_last_month'] = np.log(client['cons_last_month'] + 1)

# Plotting the boxplots for log-transformed consumption in the last year and month
plt.figure(figsize=(10, 6))
sns.boxplot(data=client[['log_cons_12m', 'log_cons_last_month']])
plt.title('Boxplot of Log-Transformed Consumption Data')
plt.xlabel('Consumption Variables')
plt.ylabel('Log of Consumption')
plt.xticks([0, 1], ['Last Year', 'Last Month'])  # Set the x-tick labels
plt.show()
```

Analysis from the Box plot:
Central Tendency and Spread: The boxes represent the interquartile range (IQR) of the data, showing the middle 50% of values. The line within each box is the median. The 'Last Year' data has a higher median compared to the 'Last Month', indicating higher median consumption annually than monthly.

Variability: The vertical lines (whiskers) extending from the boxes indicate the range of the data, excluding outliers. The 'Last Year' consumption has a wider range, suggesting more variability in annual consumption.

Outliers: The points below and above the whiskers are outliers. Both distributions have outliers, but the 'Last Year' has a more pronounced presence of lower outliers, indicating some exceptionally low consumption values.

Skewness: The 'Last Year' boxplot is skewed towards lower values (left-skewed when considering the original, non-log-transformed data), while the 'Last Month' distribution appears more symmetric around the median on the log scale

```python
forecast = client[
    ["id", "forecast_cons_12m",
    "forecast_cons_year","forecast_discount_energy","forecast_meter_rent_12m",
    "forecast_price_energy_off_peak","forecast_price_energy_peak",
    "forecast_price_pow_off_peak","churn"
    ]
]
```

```python
def plot_distribution(dataframe, column, ax):
    """Plot the distribution of a dataframe column on the provided axes."""
    sns.histplot(dataframe[column], bins=30, kde=False, ax=ax)
    ax.set_title('Distribution of ' + column)
    ax.set_ylabel('Frequency')
    ax.set_xlabel(column)

fig, axs = plt.subplots(nrows=7, figsize=(18, 40))

# Plot histogram
plot_distribution(client, "forecast_cons_12m", axs[0])
plot_distribution(client, "forecast_cons_year", axs[1])
plot_distribution(client, "forecast_discount_energy", axs[2])
plot_distribution(client, "forecast_meter_rent_12m", axs[3])
plot_distribution(client, "forecast_price_energy_off_peak", axs[4])
plot_distribution(client, "forecast_price_energy_peak", axs[5])
plot_distribution(client, "forecast_price_pow_off_peak", axs[6])
```
